```python
def data_setting(data, drop_cols, label_col):
    label = pd.DataFrame(data, columns=label_col)
    label = label[label_col[0]].to_numpy()
    
    data = data.drop(drop_cols, axis=1)
    data = data.drop(label_col, axis=1)
    return data, label

def merge(whole, x, y, dict_cols):
    data = pd.concat([x, y], axis = 1)
    join = pd.merge(data, whole, left_on=dict_cols['join'], right_on=dict_cols['join'], how='left')
    data, label = data_setting(join, dict_cols['drop'], dict_cols['label'])
    return data, label 

dict_cols = {'join':['PERSON_ID', 'HYPERTENSION'],'drop':['PERSON_ID'],'label':['HYPERTENSION']}
train_x, train_y = merge(final_k, train_x_, train_y_, dict_cols)
test_x, test_y = merge(final_k, test_x_, test_y_, dict_cols)
```
